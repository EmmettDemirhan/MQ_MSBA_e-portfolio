### Creating new feature from `amenities`

The amenities column contains a list of additional features in the property, i.e. whether it has a TV or parking. For example, a property might have the following amenities: "TV", "Wifi", "Air conditioning", "Parking", "Balcony" etc.

We will create new features from this column. For each amenity, we will create a new column in the DataFrame. For example, for the 'TV' amenity, we will create a new column called 'TV'. If a property has a TV, the value in this column will be 1, else it will be 0.

To achieve this, we first need to create a set of all possible amenities.

```python
# Example of amenities listed
df_train.amenities[:1].values
```

```python
# Creating a set of all possible amenities
# List of the relevant datasets
datasets = [df_train, df_test]

# Perform operation on each dataset
for dataset in datasets:
    amenities_list = list(dataset.amenities)
    amenities_list_string = " ".join(amenities_list)
    amenities_list_string = amenities_list_string.replace('[', '')
    amenities_list_string = amenities_list_string.replace(']', ',')
    amenities_list_string = amenities_list_string.replace('"', '')
    amenities_set = [x.strip() for x in amenities_list_string.split(',')]
    amenities_set = set(amenities_set)
amenities_set 
```

Certain amenities hold more significance than others in terms of their impact on price, such as a balcony being more likely to increase the price compared to a fax machine. Additionally, there are certain amenities that are less commonly found, like an electric profiling bed.

The block of code down is mapping amenities in the dataset to more simplified feature names, represented in a dictionary. This is done to convert these categorical amenities into binary features. Each amenity in a listing is checked against this dictionary, creating new binary columns in the dataframe, indicating the presence (1) or absence (0) of that amenity in a listing. This process is done for both the training and test datasets.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
# Scatter plot of the geographical data
plt.figure(figsize=(10, 8))
plt.scatter(df_train['longitude'], df_train['latitude'], alpha=0.2)
plt.title('Geographical data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Find optimal number of clusters
sil = [] # silhouette score placeholder
K = range(30, 50) # iterate from 2 to 15 clusters
for k in K:
    kmeans = KMeans(n_clusters=k).fit(df_train[['latitude', 'longitude']])
    preds = kmeans.predict(df_train[['latitude', 'longitude']])
    sil.append(silhouette_score(df_train[['latitude', 'longitude']], preds))

# Plot K against Silhouette score to find optimal number of clusters
plt.figure(figsize=(10, 8))
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Define the maximum number of clusters you want to check for
max_clusters = 50

# Prepare an empty list to store the inertia for each number of clusters
inertia = []

# Calculate inertia for different numbers of clusters
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_train[['latitude', 'longitude']])
    inertia.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(10, 8))
plt.plot(range(1, max_clusters + 1), inertia, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

**Determining the Optimal Number of Clusters**
To create clusters based on the geographical data, we use the KMeans algorithm from the sklearn library. To determine the optimal number of clusters, we use the silhouette score.

First, I calculate the silhouette score for a range of number of clusters (here we use 30 to 50). The silhouette score is a measure of how similar an object is to its own cluster compared to other clusters. A higher silhouette score indicates a better-defined cluster structure.But, we will go for the one perfoms better on test data which **37** weidly close to number of suburbs.

Next, we calculate the inertia for a range of number of clusters. Inertia measures the total distance of data points from their respective cluster centers. We look for the "elbow" in the plot where the decrease in inertia slows significantly. This gives us a good estimate of the optimal number of clusters.

**Clustering Geographical Data**
After determining the optimal number of clusters, we create a KMeans object and fit it to our latitude and longitude data. The result is a prediction of the cluster for each observation.

```python
# Create a KMeans object
kmeans = KMeans(n_clusters=37, random_state=0)

# Fit the KMeans object to the latitude and longitude columns and predict the cluster for each observation
clusters = kmeans.fit_predict(df_train[['latitude', 'longitude']])

# Add the clusters to the dataframe
df_train['cluster'] = clusters

# Do the same for the test set
clusters_test = kmeans.predict(df_test[['latitude', 'longitude']])
df_test['cluster'] = clusters_test
```

### Model Selection Without Hyperparameter Tuning
We first run multiple models without hyperparameter tuning. This gives us a baseline model with default parameters to compare our tuned models against. We define our target variable as 'price' and use all other numerical variables as features. We then split our training data into a training set and a validation set.

We define a dictionary of different models and fit each model on the training data, and evaluate it on the validation data. The performance metric used here is the Root Mean Square Error (RMSE), which gives us an absolute measure on how our model’s predictions deviate from the actual values.

```python
# Define your target variable (what you want to predict)
target = 'price'

# Define your features (all other columns that are not the target)
features = [col for col in df_train.columns if col != target]

# Split your training data into training and validation set
X_train, X_val, y_train, y_val = train_test_split(df_train[features],
                                                  df_train[target], test_size=0.1, random_state=42)

# Create a dictionary of different models
models = {'Linear Regression': LinearRegression(), 
          'Decision Tree': DecisionTreeRegressor(random_state=42),
          'Random Forest': RandomForestRegressor(random_state=42),
          'Gradient Boosting': GradientBoostingRegressor(random_state=42),
          'Support Vector Machines': SVR(),
          'XGBoost': XGBRegressor(random_state=42),
          'LightGBM': LGBMRegressor(random_state=42),
          'AdaBoost': AdaBoostRegressor(random_state=42)}

# Initialize a dictionary to hold your results
results = {}
# Fit each model and evaluate it
for model_name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
    results[model_name] = sqrt(mean_squared_error(y_val, predictions))

# Determine which model was the best
best_model_name = min(results, key=results.get)
best_model = models[best_model_name]

print(f'The best model was {best_model_name} with RMSE {results[best_model_name]}')

# Now fit the best model to all of the training data
best_model.fit(df_train[features], df_train[target])
```

To potentially improve our model, we perform hyperparameter tuning on the best model identified in the previous step. We use GridSearchCV, which exhaustively tries every combination of provided hyperparameters to find the best model.

The model's performance, however, did not significantly improve. This could be due to various reasons such as the model's limitations in capturing complex patterns, or the hyperparameters selected for tuning did not include the optimal values.
